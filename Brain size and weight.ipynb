```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split    
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score    
```

```python
df = pd.read_csv('Data Sets\Brain size and weight.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df[['Head Size(cm^3)']], df['Brain Weight(grams)'], color='blue')
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')  
plt.show()
```

```python
X = df[["Head Size(cm^3)"]]
y = df["Brain Weight(grams)"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
print(X_train.shape) 
print(X_test.shape) 
print(y_train.shape)
print(y_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
predict = LinearRegression()
predict.fit(X_train,y_train)
print("Model successfully trained!")
```

```python
y_pred = predict.predict(X_test)
```

```python
print(y_pred)
```

```python
y_pred = predict.predict(X_test)

score = r2_score(y_test, y_pred)

print(f"R-squared Score: {score:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='green', label='Training Data')


plt.plot(X, predict.predict(X), color='red', linewidth=3, label='Predicted Line')

plt.title('Head size vs Brain Weight (Training Set)')
plt.xlabel('HeadSize (cm^3)')  
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.grid(True)
plt.show()
```
